```python
import fitz

pdf_files = [
    ('/content/drive/MyDrive/analytica/Practical_vedanta.pdf', '/content/output_practical_vedanta.txt'),
    ('/content/drive/MyDrive/analytica/abridge_autobiography_hindi.pdf', '/content/output_autobiography_hindi.txt'),
    ('/content/drive/MyDrive/analytica/godan.pdf', '/content/output_godan.txt'),
    ('/content/drive/MyDrive/analytica/karma_yoga.pdf', '/content/output_karma_yoga.txt'),
    ('/content/drive/MyDrive/analytica/narayan_kwach.pdf', '/content/output_narayan_kwach.txt')
]

for pdf_path, output_txt_path in pdf_files:
    pdf_document = fitz.open(pdf_path)


    with open(output_txt_path, "w", encoding="utf-8") as txt_file:

        for page_num in range(len(pdf_document)):

            page = pdf_document.load_page(page_num)


            page_text = page.get_text("text")


            txt_file.write(page_text)
            txt_file.write("\n" + "="*80 + "\n")


    pdf_document.close()

    print(f"Text extracted and saved to {output_txt_path}")
```

**Practical Vedanta**

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

client = ChatNVIDIA(
    model="meta/llama-3.1-405b-instruct",
    api_key="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",
    temperature=0.2,
    top_p=0.7,
    max_tokens=1024,
)

with open('/content/drive/MyDrive/analytica/output_practical_vedanta.txt', 'r', encoding='utf-8') as file:
    data = file.read()

paragraphs = data.split('================================================================================')

paragraphs = [paragraph.strip() for paragraph in paragraphs if paragraph.strip()]

max_paragraphs = 6
questions = []
for i, paragraph in enumerate(paragraphs[:max_paragraphs], start=1):
    prompt = f"Based on the following English paragraph, generate only few good questions in English: {paragraph}"

    print(f"Generating questions for Paragraph {i}...")

    full_response = ""
    for chunk in client.stream([{"role": "user", "content": prompt}]):
        full_response += chunk.content

    questions.append(full_response.strip())
    print(f"Questions for Paragraph {i}:\n{full_response.strip()}\n")

with open('generated_questions.txt', 'w', encoding='utf-8') as output_file:
    for i, question in enumerate(questions, start=1):
        output_file.write(f"Questions for Paragraph {i}:\n{question}\n\n")

print("All questions have been generated and saved to 'generated_questions.txt'.")
```
